```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def log_transform(image, c=1.0):
    transformed_image = c * np.log1p(image)

    transformed_image = (255 * transformed_image / np.max(transformed_image)).astype(np.uint8)

    return transformed_image

image = cv2.imread('F:/Nitheesh/Computer vision/Exercise5/image1.jpg')

c = 1.0

transformed_image = log_transform(image, c)

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(transformed_image)
plt.title('Log Transformed Image')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np

def log_transform(image, c=1.0):
    transformed_image = c * np.log1p(image)

    transformed_image = (255 * transformed_image / np.max(transformed_image)).astype(np.uint8)

    return transformed_image

def main():
    image_path = 'F:/Nitheesh/Computer vision/Exercise5/image1.jpg'
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    c = 1.0

    transformed_image = log_transform(image, c)

    # Display the original and transformed images using OpenCV
    cv2.imshow('Original Image', image)
    cv2.imshow('Log Transformed Image', transformed_image)

    # Wait for a key press and close the windows
    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
